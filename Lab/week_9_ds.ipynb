```python
import numpy as np
from scipy.stats import bernoulli, binom

print("----- Bernoulli Distribution -----")

p = 0.7

bernoulli_samples = bernoulli.rvs(p, size=10)

print("Probability of Passing:", p)
print("Student Outcomes:", bernoulli_samples)

print("Probability of Pass:")
print(bernoulli.pmf(1, p))

print("Probability of Fail:")
print(bernoulli.pmf(0, p))

print("\n----- Binomial Distribution -----")

n = 10
k = 8

binomial_probability = binom.pmf(k, n, p)

print("Probability that exactly 8 students pass:")
print(binomial_probability)

random_binomial = binom.rvs(n, p, size=5)
print("Random Outcomes:", random_binomial)
```

```python
import numpy as np
from scipy.stats import norm

np.random.seed(42)

population = np.random.randint(50, 96, 1000)

sample_size = int(0.10 * len(population))
sample = np.random.choice(population, size=sample_size, replace=False)

population_mean = np.mean(population)
population_std = np.std(population)

sample_mean = np.mean(sample)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - norm.cdf(abs(z_score)))

print("Population Mean:", population_mean)
print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
import numpy as np
from scipy import stats

class_A = np.array([78, 82, 85, 88, 90])
class_B = np.array([72, 75, 70, 68, 74])

t_stat, p_value = stats.ttest_ind(class_A, class_B)

print("t-Statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
import numpy as np
from scipy import stats

group1 = np.array([45, 50, 52, 55, 58, 60, 62, 65])
group2 = np.array([48, 53, 57, 59, 61, 64, 67, 70])

u_stat, p_value = stats.mannwhitneyu(group1, group2)

print("U-statistic:", u_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
import numpy as np
from scipy import stats

sample = np.random.uniform(0, 1, 100)

ks_stat, p_value = stats.kstest(sample, 'uniform')

print("KS Statistic:", ks_stat)
print("P-value:", p_value)
```
